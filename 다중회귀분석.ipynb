```python
'''
상영시간 / 장르 / 제작국가 / 심의 / 관람등급에 따른 누적 매출액과 누적 관객수
다중회귀분석 진행

1. movie.csv를 읽어가면서 처음 나오는 것들에 대한 정보 저장
2. movie_cd를 꺼내서 영화상세정보 갖고오기
3. pandas dataframe형식으로 설정하기
4. 다중회귀분석진행
5. 결과분석
'''
import statsmodels.api as sm
import pandas as pd
import requests 
import re
import json
key = 'd6700ce7dc21e4b608ddcf1bcb4b2269'
'''
movieCd : 영화코드
movieNm : 영화이름
salesAcc : 누적매출액
audiAcc : 누적관객수
'''
mv = pd.read_csv('movie.csv',encoding='ms949')

''' 중복 제거하고 누적매출액 누적 관객수 제일 많은 상태로 저장 '''

code_list = set([])
sales_list = []
audi_list = []
for code,name,sales,audi in zip(mv['movieCd'],mv['movieNm'],mv['salesAcc'],mv['audiAcc']): 
    if code not in code_list:
        code_list.add(code)
        sales_list.append(sales)
        audi_list.append(audi)

''' 누적 관객수와 누적 매출액 사이의 회귀분석 진행'''        
body = pd.DataFrame(
    {
        'sales' : sales_list,
        'audi'  : audi_list
    }
)
reg = sm.OLS.from_formula("sales ~ audi",body).fit()
reg.summary()
```
